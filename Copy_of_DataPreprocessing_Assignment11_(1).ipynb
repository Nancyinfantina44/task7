**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Step 2: Importing dataset**

```python
df=pd.read_csv("/content/Data.csv")
df
```

**Step 3: Handling the missing data**

```python
df.isna().sum()
```

```python
df.describe()
```

```python
mean_value=df['Age'].mean().round()
df['Age'].fillna(value=mean_value,inplace=True)
#df
```

```python
df
```

```python
means=df['Salary'].mean()
df['Salary'].fillna(value=means,inplace=True)
df
```

```python
df['Salary']=df['Salary'].round()
df
```

**Step 4: Encoding categorical data**

```python
df.Country.unique()
```

```python
df['Purchased']=df['Purchased'].map({'Yes':1,'No':0})
df
```

```python
df['Country']=df['Country'].map({'France':1,'Spain':2,'Germany':3})
df
```

**Step 5: Creating a dummy variable**

```python
pd.get_dummies(df)
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
from sklearn.model_selection import train_test_split
x=df[['Country','Age','Salary']].values
y=df['Purchased'].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train1=scaler.transform(x_train)
x_test1=scaler.transform(x_test)
```
